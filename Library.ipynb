### Tables File - Uses

* Available uses: `wr_usage`, `wr_performance`, `rb_usage`, `rb_performance`, `te_usage`, `te_performance`, `qb_usage`, `qb_performance`
* Available inputs for all of these are the exact same
  - Week: Integer
  - Sort: Which column you'd like to sort by. Best practice is to just pull up the chart then add which column you'd like to sort by. It's a string input. Put in wr_usage(1, sort='column_to_sort_by').
  - Amount is an integer. Default depends on position. Generally around 30.
  - Style: default is none and the output is a colored table. If you want the raw data just put style=''.
* wr_usage(1, sort='W1 Points', amount=40, style='')
  - (week, sort by week 1 fantasy points, amount of players, style's output is a raw dataframe in this case.

```python
tb.wr_usage(2, 'W2 AirYds', amount=30)
```

```python
tb.wr_performance(2, amount=30)
```

```python
tb.rb_usage(2, amount=20)
```

```python
tb.rb_performance(2, sort='W2 YPC', amount=20)
```

```python
tb.te_usage(2, amount=12)
```
